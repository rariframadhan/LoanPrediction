### Cek Persebaran Data

Data Numerical

```python
data_num = data.select_dtypes(include ="number")
len(data_num.columns)
```

```python
data_num = data_num
k = len(data_num.columns)
n = 3
m = (k - 1) // n + 1
fig, axes = plt.subplots(m, n, figsize=(n * 5, m * 3))
for i, (name, col) in enumerate(data_num.iteritems()):
    r, c = i // n, i % n
    axe = axes[r, c]
    sns.kdeplot(x = col, color='green', data=data_num, ax=axe)

fig.tight_layout()
```

in data distribution we know mayority data have skew distribution
"Target" is target

```python
plt.figure(figsize=(15,10))
sns.barplot(x ="AMT_INCOME_TOTAL" ,y= "SK_ID_CURR", data=data_num)
plt.xlabel(rotate = 90)
```
