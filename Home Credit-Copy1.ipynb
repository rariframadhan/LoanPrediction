```python
data_all = data_analysis.replace({"NAME_CONTRACT_TYPE":{"Cash loans":0 , "Revolving loans":1} , "CODE_GENDER":{"M":0,"F":1, "XNA":1},
                                 "FLAG_OWN_CAR":{"N":0,"Y":1}, "FLAG_OWN_REALTY":{"N":0,"Y":1} , "NAME_CONTRACT_STATUS" : {"Approved": 1, "Refused" : 0, "Canceled": 0, "Unused offer" : 0 } ,
                                 "CREDIT_ACTIVE" : {"Closed": 0, "Active" : 1, "Sold": 0, "Bad debt" : 1 } }) 
data_all.head()
```

```python
#Make Phone ownership
def telepon(data):
    if data["FLAG_EMP_PHONE"] == 1 or data["FLAG_WORK_PHONE"] == 1 or data["FLAG_CONT_MOBILE"] == 1 or data["FLAG_PHONE"] == 1 or data["FLAG_MOBIL"] == 1:
        return 1
    else :
        return 0
```

```python
data_all["PHONE"] = data_all.apply(telepon , axis = 1)
data_all.drop(columns=["FLAG_EMP_PHONE","FLAG_WORK_PHONE","FLAG_CONT_MOBILE","FLAG_PHONE"] , inplace=True)
```

```python
#Accompanied
def accompanied(data):
    if data == "Unaccompanied":
        return 0
    else:
        return 1
```

```python
data_all.NAME_TYPE_SUITE = data_all.NAME_TYPE_SUITE.apply(accompanied)
```

```python
#Marriage
def marriage(data):
    if data == "Married" or data == "Civil marriage":
        return 1
    else:
        return 0
```

```python
data_all.NAME_FAMILY_STATUS = data_all.NAME_FAMILY_STATUS.apply(marriage)
```

```python
#Working
def working(data):
    kerja = ["Working", "State servant", "Commercial associate"]
    if data in kerja:
        return 1
    else :
        return 0
```

```python
data_all["NAME_INCOME_TYPE"] = data_all.NAME_INCOME_TYPE.apply(working)
```

```python
#Have House
def house(data):
    if data == "House / apartment":
        return 1
    else :
        return 0
```

```python
data_all["NAME_HOUSING_TYPE"] = data_all.NAME_HOUSING_TYPE.apply(house)
```

```python
#Have Skill Education
def edu(data):
    if data == "Academic degree" or data == "Higher education":
        return 1
    else :
        return 0
```

```python
data_all["NAME_EDUCATION_TYPE"] = data_all.NAME_EDUCATION_TYPE.apply(edu)
```

```python
# Drop Occupation Type
data_all.drop(columns = ["OCCUPATION_TYPE", "FLAG_MOBIL", "NAME_HOUSING_TYPE"], inplace = True)
```

```python
#install imblearn
##! pip install imblearn
```

```python
from collections import Counter
print(sorted(Counter(data_all["Status"]).items()))
```

```python
#split data
X = data_all.drop(columns = ["Id","Status"])
Y = data_all["Status"]
```

```python
#undersampling
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=0)
x_resampler , y_resampler = rus.fit_resample(X,Y)
print(sorted(Counter(y_resampler).items()), y_resampler.shape)
```

### Separated Data

```python
X = x_resampler
Y = y_resampler
```

```python
# membagi dataframe menjadi data train dan data test
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 42)
```

### Modeling

Logistic Regression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve, auc

def eval_classification(model, pred, xtrain, ytrain, xtest, ytest):
    print("Accuracy (Test Set): %.2f" % accuracy_score(ytest, pred))
    print("Precision (Test Set): %.2f" % precision_score(ytest, pred))
    print("Recall (Test Set): %.2f" % recall_score(ytest, pred))
    print("F1-Score (Test Set): %.2f" % f1_score(ytest, pred))
    
    fpr, tpr, thresholds = roc_curve(ytest, pred, pos_label=1) # pos_label: label yang kita anggap positive
    print("AUC: %.2f" % auc(fpr, tpr))
def show_feature_importance(model):
    feat_importances = pd.Series(model.feature_importances_, index=X.columns)
    ax = feat_importances.nlargest(25).plot(kind='barh', figsize=(10, 8))
    ax.invert_yaxis()

    plt.xlabel('score')
    plt.ylabel('feature')
    plt.title('feature importance score')

def show_best_hyperparameter(model, hyperparameters):
    for key, value in hyperparameters.items() :
        print('Best '+key+':', model.get_params()[key])
```

```python
# do the modeling
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)
model.fit(X_train, Y_train)
```

Bandingkan Hasil

```python
#predict
Y_pred = model.predict(X_test)
Y_pred
```

```python
#train
Y_pred_train = model.predict(X_train)
Y_pred_train
```

Evaluation

```python
eval_classification(model, Y_pred, X_train, Y_train, X_test, Y_test)
```

```python
print("Precision (Train Set): %.2f" % precision_score(Y_train, Y_pred_train))
```
